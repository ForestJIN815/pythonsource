```python
# score.txt 에 있는 점수를 읽어 합계와 평균 구한 후 출력

f = open("./score.txt", "r", encoding="utf-8")
scores = []
for num in f:
    scores.append(int(num))

print(f"평균 : {sum(scores) / len(scores)}")
```

```python
# sample.txt 읽은 후 합계와 평균 구하기
# result.txt 작성(합계, 평균 기록)

scores = []
f = open("./sample.txt", "r", encoding="utf-8")
for num in f:
    scores.append(int(num))

f1 = open("./result.txt", "w", encoding="utf-8")
f1.write(f"합계 : {sum(scores)}\n")
f1.write(f"평균 : {sum(scores)/len(scores)}\n")
```

```python
# info.txt 읽기
# BMI 지수 계산 후 화면에 출력
# [출력결과]
# 사라, 59, 180
# 이름 : 사라
# 몸무게 : 59
# 키 : 180
# BMI : 18.209 (bmi = weight / (height / 100) ** 2)
# 결과 : 저체중
f = open("./info.txt", "r", encoding="utf-8")
for info in f:
    p = info.strip()
    print(p)
    name, weight, height = info.strip().split(", ")
    # print(name, weight, height)
    bmi = int(weight) / (int(height) / 100) ** 2
    
    print(f"이름 : {name}\n")
    print(f"몸무게 : {weight}\n")
    print(f"키 {height}\n")
    print(f"BMI {bmi}\n")

    if 25 <= bmi:
        result = "과체중"
    elif 18.5 <= bmi:
        result = "정상체중"
    else:
        result = "저체중"
    
    print(name, weight, height)
    print(f"이름 : {name}")
    print(f"몸무게 : {weight}")
    print(f"키 : {height}")
    print(f"BMI : {bmi}")
    print(f"결과 : {result}")

    # 위아래 띄기? 췍
```

```python
while True:
    no = int(input("1. 암호화 | 2. 복호화 | 3. 종료 "))
    if no == 1:
        # origin 읽은 후 암호화 후 encryption.txt 로 저장
        f = open("./origin.txt", "r", encoding="utf-8")
        content = f.read()

        f1 = open("./encryption.txt", 'w', encoding="utf-8")
        for c in content:
            f1.write(chr(ord(c) + 100))

    elif no == 2:
        # encryption.txt 읽은 후 원래 문자대로 화면 출력
        f2 = open("./encryption.txt", 'r', encoding="utf-8")
        content = f2.read()
        for i in range(len(content)):
            print(chr(ord(content[i]) - 100), end="")
    else:
        break ## 답이 안나옴, 췍
```

```python
# close() 자동으로 호출해줌

with open("./origin.txt", "r", encoding="utf-8") as f1:
    print(f1.read())
```
