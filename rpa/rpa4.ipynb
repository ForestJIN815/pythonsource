```python
# 스크롤

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup

driver = webdriver.Chrome()
driver.maximize_window()

driver.get("https://shopping.naver.com/ns/home")
time.sleep(1)

# 상품명 입력
element = driver.find_element(By.CLASS_NAME,"_searchInput_search_text_83jy9")
element.send_keys("마우스")
element.send_keys(Keys.ENTER)
time.sleep(3)

# 페이지 이동 후 스크롤 이동 : window.scrollTo(시작위치, 종료위치)
# driver.execute_script("window.scrollTo(0, 2160)")

# 현재 문서 높이
prev_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # 스크롤 이동
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    time.sleep(2)

    # 스크롤이 진행된 후 문서 높이
    curr_height = driver.execute_script("return document.body.scrollHeight")

    if curr_height == prev_height:
        break

    prev_height = curr_height

# 맨 처음으로 가기
driver.execute_script("window.scrollTo(0, 0)")


time.sleep(3)

driver.quit()
```

```python
# 네이버 이미지 다운
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
from urllib.request import urlretrieve

driver = webdriver.Chrome()
driver.maximize_window()

driver.get("https://www.naver.com")
time.sleep(1)

element = driver.find_element(By.ID, "query")
element.send_keys("바이올린")
element.send_keys(Keys.ENTER)

time.sleep(2)

# 이미지 메뉴 클릭
driver.find_element(By.XPATH, '//*[@id="lnb"]/div[1]/div/div[1]/div/div[1]/div[1]/a').click()
time.sleep(2)

prev_height = driver.execute_script("return document.body.scrollHeight")
# 스크롤 스크립트
while True:
    # 스크롤 이동
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    time.sleep(2)

    # 스크롤이 진행된 후 문서 높이
    curr_height = driver.execute_script("return document.body.scrollHeight")

    if curr_height == prev_height:
        break

    prev_height = curr_height

# 맨 처음으로 가기
driver.execute_script("window.scrollTo(0, 0)")

time.sleep(3)

# 이미지 요소들 가져오기
images = driver.find_elements(By.CSS_SELECTOR, ".thumb .link_thumb")

count = 1
for img in images:
    img.click()
    time.sleep(2)

    # 오른쪽에 뜬 큰 이미지 저장
    large_img = driver.find_element(By.CSS_SELECTOR, ".viewer_image img").get_attribute("src")
    print(large_img)

    urlretrieve(large_img, "./download/"+str(count)+".png")
    count += 1


time.sleep(3)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from openpyxl import Workbook
from openpyxl.drawing.image import Image
import requests
from fake_useragent import UserAgent
from io import BytesIO

wb = Workbook()
ws = wb.active
ws.column_dimensions["B"].width = 70
ws.column_dimensions["C"].width = 20
ws.append(['번호','제품명','가격','이미지'])

driver = webdriver.Chrome()
driver.maximize_window()

driver.get("https://prod.danawa.com/list/?cate=112758&15main_11_02=")
time.sleep(2)

# 제조사별 클릭하기
driver.find_element(By.XPATH, '//*[@id="dlMaker_simple"]/dd/div[2]/button[1]').click()
time.sleep(1)

# 애플 클릭하기
driver.find_element(By.XPATH, '//*[@id="selectMaker_simple_priceCompare_A"]/li[15]').click()
time.sleep(2)


cur_page, target_crawl_num = 1,5
idx= 1

while cur_page <= target_crawl_num:

    print(f'현재 페이지 : {cur_page}')

    # 데이터 수집
    # 제품명, 가격(첫번째), img src
    product_list = driver.find_elements(By.CSS_SELECTOR, ".product_list .prod_item")
    for product in product_list:
        title = product.find_element(By.CSS_SELECTOR, ".prod_name a").text
        img_src = product.find_element(By.CSS_SELECTOR, ".thumb_image img")
     
        if img_src.get_attribute("data-original"):
            src = img_src.get_attribute("data-original")
        else:
            src = img_src.get_attribute('src')

        if "https:" not in src:
            src = "https:"+src

        price = product.find_element(By.CSS_SELECTOR, ".prod_pricelist p.price_sect strong").text

        print(idx, title, price, src)
        # 번호 제품명 가격 이미지
        ws.append([idx, title, price])
        idx += 1

        # 이미지 가져와서 엑셀 저장
        res = requests.get(src,headers={"user-agent":UserAgent().chrome})
        save_img = BytesIO(res.content) # content : byte 로 처리

        img = Image(save_img)
        img.width = 30
        img.height = 20
        ws.add_image(img, "D"+str(idx))

    cur_page += 1

    if cur_page > target_crawl_num:
        break

    # 다음페이지 번호 클릭하기
    # div.number_wrap > a:nth-child(2)    
    # div.number_wrap > a:nth-child(3)
    # div.number_wrap > a:nth-child(4)
    # div.number_wrap > a:nth-child(5)
    driver.find_element(By.CSS_SELECTOR, f"div.number_wrap > a:nth-child({cur_page})").click()
    time.sleep(3)

wb.save("./danawa.xlsx")


driver.quit()
```

```python
# db 저장

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import sqlite3
import re

conn = sqlite3.connect("./danawa.db",isolation_level=None)
cursor = conn.cursor()

driver = webdriver.Chrome()
driver.maximize_window()

driver.get("https://prod.danawa.com/list/?cate=112758&15main_11_02=")
time.sleep(2)

# 제조사별 클릭하기
driver.find_element(By.XPATH, '//*[@id="dlMaker_simple"]/dd/div[2]/button[1]').click()
time.sleep(1)

# 애플 클릭하기
driver.find_element(By.XPATH, '//*[@id="selectMaker_simple_priceCompare_A"]/li[15]').click()
time.sleep(2)


cur_page, target_crawl_num = 1,5
idx= 1

while cur_page <= target_crawl_num:

    print(f'현재 페이지 : {cur_page}')

    # 데이터 수집
    # 제품명, 가격(첫번째), img src
    product_list = driver.find_elements(By.CSS_SELECTOR, ".product_list .prod_item")
    for product in product_list:
        title = product.find_element(By.CSS_SELECTOR, ".prod_name a").text
        img_src = product.find_element(By.CSS_SELECTOR, ".thumb_image img")
     
        if img_src.get_attribute("data-original"):
            src = img_src.get_attribute("data-original")
        else:
            src = img_src.get_attribute('src')

        if "https:" not in src:
            src = "https:"+src

        price = product.find_element(By.CSS_SELECTOR, ".prod_pricelist p.price_sect strong").text

        print(idx, title, price, src)
        # db 저장
        price = re.sub(",","",price)
        sql = "INSERT INTO apple(title,price,url) VALUES(?,?,?)"
        cursor.execute(sql, (title,price,src))
      
        idx += 1        

    cur_page += 1

    if cur_page > target_crawl_num:
        break

    # 다음페이지 번호 클릭하기
    # div.number_wrap > a:nth-child(2)    
    # div.number_wrap > a:nth-child(3)
    # div.number_wrap > a:nth-child(4)
    # div.number_wrap > a:nth-child(5)
    driver.find_element(By.CSS_SELECTOR, f"div.number_wrap > a:nth-child({cur_page})").click()
    time.sleep(3)

driver.quit()
```
